```python
# Encapsulation
'''You can hide the inner details of a class or certain data you only need to share or expose what is needed for
the user of the class to use this  class

- Instead of accessing directly acess data throgh getters and setters
- Property 
'''

#Inheritance
''' Automatically allows us to have certain attributes for objects because they are defined in a base class

User
 1. Customers
 2. Teachers

'''

#Polymorphism
'''
We can treat customers and techers as the same thing if we approach them as users. We can say "Hey users we need 
you to do this " or " We need to update this data ". As a result since customer is a user that is okay.  And a teacher
is a user that is okay. But customers can do something specific. And teachers can do something specific.

'''
```

```python
class User:
    def log(self):
        print(self)


class Customer(User):
    def __init__(self, name, membership_type):
        self.Name = name
        self.Type = membership_type
        
    
    # This is an example of a decorator
    # It gives extra functionality
    @property
    def Name(self):
#         print("Getting name")
        return self._Name # This underscore is a way to define private variabels
        # Don't touch it , if it is not within the class
        # So outside of the class we should not use "custometr[0]._name"
        # When we try to get the name this function will be invoked
        # _name is the backing filed , it is the data behind the scenes
        # Any time we try to set this data we invoke another method
        # The way define that is like below
        
    @Name.setter
    def Name(self,name):
#         print("Setting name")
        self._Name = name
        
    
        
    def update_membership(self, new_membership):
        print("Updating the membership...")
        self.Type = new_membership
        
    def __str__(self):
        return self.Name + " " + self.Type
    
    def print_all_customers(customers):
        for i in customers:
            print(i)
            
    def __eq__(self,other):
        if self.Name == other.Name and self.Type == other.Type:
            return True
        else:
            return False
        
    __hash__ = None
     
    #Print all the customers with details
    __repr__ = __str__
 
     
customers = [Customer("Buddhika","Gold"), Customer("Weerasinghe","Silver") ]

print(customers[0].Name)
print("\n")
customers[0].log()
```

```python
class User:
    def log(self):
        print("Hey I am a user")

class Teacher:
    def log(self):
        print("I am a teacher")
        
class Customer(User):
    
    def log(self):
        print("I am a customer")
    
    def __init__(self, name, membership_type):
        self.Name = name
        self.Type = membership_type
        
    
    # This is an example of a decorator
    # It gives extra functionality
    @property
    def Name(self):
#         print("Getting name")
        return self._Name # This underscore is a way to define private variabels
        # Don't touch it , if it is not within the class
        # So outside of the class we should not use "custometr[0]._name"
        # When we try to get the name this function will be invoked
        # _name is the backing filed , it is the data behind the scenes
        # Any time we try to set this data we invoke another method
        # The way define that is like below
        
    @Name.setter
    def Name(self,name):
#         print("Setting name")
        self._Name = name
        
    
        
    def update_membership(self, new_membership):
        print("Updating the membership...")
        self.Type = new_membership
        
    def __str__(self):
        return self.Name + " " + self.Type
    
    def print_all_customers(customers):
        for i in customers:
            print(i)
            
    def __eq__(self,other):
        if self.Name == other.Name and self.Type == other.Type:
            return True
        else:
            return False
        
    __hash__ = None
     
    #Print all the customers with details
    __repr__ = __str__
 
# We treat them all as users and we invoke .log()
# We can override the base class in the child class
# This is polymorphism

users = [Customer("Buddhika","Gold"), Customer("Weerasinghe","Silver") , Teacher()]

for user in users:
    user.log()
```
