**Barchart**

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
import numpy as np
xpos = np.arange(len(company))
xpos
```

```python
# plt.bar(company, revenue) ''' Here we use ypos because we cant use strings inside '''
plt.xticks(xpos, company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Tech stocks")
plt.bar(xpos, revenue)
```

```python
# plt.bar(company, revenue) ''' Here we use ypos because we cant use strings inside '''
plt.xticks(xpos, company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Tech stocks")
plt.bar(xpos, revenue, label ='Revenue')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
# plt.bar(company, revenue) ''' Here we use ypos because we cant use strings inside '''
plt.xticks(xpos, company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Tech stocks")
plt.bar(xpos-0.2, revenue, label ='Revenue')
plt.bar(xpos+0.2, profit, label ='Profit') # Add profit to the chart
plt.legend()

# As we have used numpy arrays in xpos- 0.2 will create a new array with 4 elements
# And 0.2 will be substracted from each of these elements
```

```python
# plt.bar(company, revenue) ''' Here we use ypos because we cant use strings inside '''
plt.xticks(xpos, company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Tech stocks")
plt.bar(xpos-0.2, revenue,width =0.4, label ='Revenue')
plt.bar(xpos+0.2, profit, width =0.4,label ='Profit') # Add profit to the chart
plt.legend()
```

```python
# plt.bar(company, revenue) ''' Here we use ypos because we cant use strings inside '''
plt.yticks(xpos, company)
plt.ylabel("Company")
plt.xlabel("Revenue")
plt.title("US Tech stocks")


plt.barh(xpos-0.2, revenue, label ='Revenue')
plt.barh(xpos+0.2, profit,label ='Profit') # Add profit to the chart
plt.legend()
```

**Histograms**

* Histograms are bar charts where the y axis contains the frequency of a variable

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins is set to 10
# rwidth is a relative width of a bar compared to a bin size
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = 3) 
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = 3,rwidth=0.8) 
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = [80,100,125,150],rwidth=0.8) 
# 3 bins
# 80 -100
# 100 - 125
# 125 -150
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = [80,100,125,150],rwidth=0.95, color ='g') 
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = [80,100,125,150],rwidth=0.95, color ='g', histtype ='step') 
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

# Here we make the first argument an array
plt.xlabel('sugar range')
plt.ylabel('Total no of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,150],rwidth=0.95, color =['green','orange'],
        label = ['men', 'women']) 
plt.legend()
```

**PIE CHART**

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 2)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.2f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.1f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%')
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%', shadow = True )
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%', shadow = True, explode =
       [0,0.5,0,0,0])
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%', shadow = True, explode =
       [1,0.5,0.3,0.2,0.1])
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%', shadow = True, explode =
       [0,0.1,0.1,0,0], startangle = 45)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%', shadow = True, explode =
       [0,0.1,0.1,0,0], startangle = 180)
plt.show()
```

**Save chart to a file using savefig**

```python
plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%', shadow = True, explode =
       [0,0.1,0.1,0,0], startangle = 180)
plt.savefig("piechart.png")
```

```python
# Perfectly fit chart into a image

plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%', shadow = True, explode =
       [0,0.1,0.1,0,0], startangle = 180)
plt.savefig("piechart.png",bbox_inches = 'tight')
```

```python
# Have a padding around the image

plt.pie(exp_vals,labels=exp_labels, radius = 1.5, autopct = '%0.0f%%', shadow = True, explode =
       [0,0.1,0.1,0,0], startangle = 180)
plt.savefig("piechart.png",bbox_inches = 'tight', pad_inches = 2)
```
